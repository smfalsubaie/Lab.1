```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

# Step size
h = 0.1
x = np.arange(0, 2*np.pi, h)
y = np.cos(x)

# Forward difference approximation
forward_diff = np.diff(y)/h
x_diff = x[:-1]
exact_solution = -np.sin(x_diff)

# Plot
plt.figure(figsize=(12, 8))
plt.plot(x_diff, forward_diff, '--', label='Finite difference approximation')
plt.plot(x_diff, exact_solution, label='Exact solution')
plt.legend()
plt.show()

# Max error
max_error = max(abs(exact_solution - forward_diff))
print("Maximum error:", max_error)
```

```python
import numpy as np

# Point at which to compute the derivative
x0 = 0.7

# Step sizes: h = 2^(-1), 2^(-2), ..., 2^(-29)
h = 2.0 ** -np.arange(1, 30)

# Forward finite difference approximation
df = (np.cos(x0 + h) - np.cos(x0)) / h

# True derivative value at x0
true_value = -np.sin(x0)

# Print table header
print("k | Approximation         | Ratio of errors     | Relative differences")
print("---|------------------------|---------------------|-------------------------")

previous_approximation = None
previous_error = None

for k in range(1, len(h) + 1):
    approximation = df[k - 1]
    error = abs(approximation - true_value)

    # Compute ratio of errors (previous / current)
    ratio = abs(previous_error / error) if previous_error is not None else " "

    # Compute relative difference between successive approximations
    relative_diff = abs((approximation - previous_approximation) / previous_approximation) \
        if previous_approximation is not None else " "

    print(f"{k:<2} | {approximation:<22.15f} | {ratio if isinstance(ratio, str) else f'{ratio:<19.6f}'} | {relative_diff if isinstance(relative_diff, str) else f'{relative_diff:<23.10f}'}")

    previous_approximation = approximation
    previous_error = error

# Print the true derivative value
print(f"\nTrue value of the derivative: {true_value:.17f}")
```
